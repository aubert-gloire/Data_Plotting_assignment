```python
# Set the style
sns.set(style="whitegrid")

# Create the histogram
plt.figure(figsize=(12, 6))
sns.histplot(df['co2'], bins=30, kde=True, color='blue')

# Label the chart
plt.title('Distribution of Carbon Dioxide Emissions (CO2)', fontsize=16)
plt.xlabel('CO2 Emissions (g/mile)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)

# Show the plot
plt.show();
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['UCity'], bins=30, kde=True, color='#4A90E2')  # bright blue

plt.title('Distribution of Urban Fuel Economy (UCity)', fontsize=16, color='#1A237E')  # dark blue
plt.xlabel('Urban Fuel Economy (mpg)', fontsize=12, color='#0D47A1')
plt.ylabel('Frequency', fontsize=12, color='#0D47A1')
plt.grid(True, color='#BBDEFB')  # light blue grid

plt.show()



```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['cylinders'], bins=range(int(df['cylinders'].min()), int(df['cylinders'].max()) + 2), color='#E57373')  # soft red

plt.title('Distribution of Engine Cylinders', fontsize=16, color='#B71C1C')  # dark red
plt.xlabel('Number of Cylinders', fontsize=12, color='#C62828')
plt.ylabel('Frequency', fontsize=12, color='#C62828')
plt.grid(True, color='#FFCDD2')  # light pink grid

plt.show()
```

```python
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Set plot
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', center=0, linewidths=0.5, fmt=".2f")

# Title and adjustments
plt.title('Correlation Heatmap of Fuel Economy Dataset', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

plt.show()

```

# Data intepretation

1.Strong Positive Correlations (Values close to +1) These indicate that when one variable increases, the other also tends to increase.

City, UCity, Highway, UHighway, and Comb: These are all strongly correlated (> 0.9) with each other, meaning different measures of fuel economy move together.



co2 vs. cylinders/displ: Strong positive correlation (~0.85–0.86), indicating that larger engines produce more CO2 emissions.

feScore and ghgScore: Very strongly positively correlated with each other (0.99), which makes sense since both relate to vehicle efficiency.

2.Strong Negative Correlations (Values close to -1) These indicate that as one variable increases, the other tends to decrease.

Fuel efficiency (city, highway, comb, etc.) vs. CO2: Very strong negative correlation (~-0.9), which shows that more efficient cars emit less CO2.

Fuel efficiency vs. cylinders and displacement (displ): Also strongly negative (~-0.7 to -0.8), indicating that larger engines are less fuel efficient.

feScore and ghgScore vs. engine size (displ/cylinders): Around -0.78 to -0.79, reinforcing that bigger engines are worse for the environment..

3.Interesting Observations pv2 and pv4 (perhaps powertrain type indicators) have a negative correlation with each other (-0.67), which might imply mutual exclusivity.

Year: Weak correlation with most variables, suggesting that tech improvements vary and don’t follow a simple trend over time.
